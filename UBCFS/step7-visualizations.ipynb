```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data/final/2022_2023_CFFS_Outcomes/UBCFS_summary.csv").drop(["Unnamed: 0"], axis=1)
data
```

```python
gather = pd.read_csv("data/final/2022_2023_CFFS_Outcomes/Data_Labelled_Gather22-23_with_name.csv")
gather["Restaurant"] = "Gather"

ok = pd.read_csv("data/final/2022_2023_CFFS_Outcomes/Data_Labelled_OK22-23_with_name.csv")
ok["Restaurant"] = "Open Kitchen"

feast = pd.read_csv("data/final/2022_2023_CFFS_Outcomes/Data_Labelled_Feast22-23_with_name.csv")
feast["Restaurant"] = "Feast"

df = pd.concat([gather, ok], axis=0)
df = pd.concat([feast, df], axis=0)
df
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Table(header=dict(values=["ProdId", "Weight (g)", "GHG Emission (g) / 100g", "N lost (g) / 100g",
                                                  "Stress-Weighted Water Use (L) / 100g","Combined Label"],
                                          fill_color="lightskyblue", font=dict(color="black", size=12)),
                              cells=dict(values=[df["ProdId"].tolist(), df["Weight (g)"].tolist(), 
                                                df["GHG Emission (g) / 100g"].tolist(), df["N lost (g) / 100g"].tolist(),
                                                df["Stress-Weighted Water Use (L) / 100g"].tolist(),
                                                df["Combined Label"].tolist()],
                                        fill_color="white", align="left", line_color="lightgrey"))])

fig.show()
fig.write_html("result_table.html")
```

```python
ok_df = data.loc[data["Restaurant"] == "Open Kitchen"]
feast_df = data.loc[data["Restaurant"] == "Feast"]
gather_df = data.loc[data["Restaurant"] == "Gather"]
print(len(ok_df))
print(len(feast_df))
print(len(gather_df))

fig, axes = plt.subplots(1,3, figsize=(12,6), sharey=True)

palette = {"Green":"tab:green", "Red":"tab:red", "Yellow":"tab:orange"}

ax1 = sns.countplot(data=ok_df, x="Label", ax=axes[0], palette=palette)
ax1.set_title("Open Kitchen")
ax1.set_xlabel("")
ax2 = sns.countplot(data=feast_df, x="Label", ax=axes[1], palette=palette)
ax2.set_title("Gather")
ax2.set_xlabel("")
ax2.set_ylabel("")
ax3 = sns.countplot(data=gather_df, x="Label", ax=axes[2], palette=palette)
ax3.set_title("Feast")
ax3.set_xlabel("")
ax3.set_ylabel("")
for ax in [ax1, ax2, ax3]:
    for i in ax.containers:
        ax.bar_label(i, )
        
ax1.set_title("Open Kitchen")
ax2.set_title("Feast")
ax3.set_title("Gather")
plt.savefig("data/final/2022_2023_CFFS_Outcomes/UBCFS_Summary_fig.png")
plt.tight_layout()
```

```python
red = len(data.loc[data["Label"] == "Red"])
green = len(data.loc[data["Label"] == "Green"])
yellow = len(data.loc[data["Label"] == "Yellow"])

label = ["Red", "Green", "Yellow"]
label_data = [red, green, yellow]

palette = {"Green":"tab:green", "Red":"tab:red", "Yellow":"tab:orange"}

plt.figure(figsize=(8,7))
plt.pie(label_data, labels=label, colors=[palette[key] for key in label], autopct='%1.1f%%', textprops={'fontsize': 14})
plt.title("Food Labeling Results Across All Restaurants")
plt.savefig("data/final/2022_2023_CFFS_Outcomes/UBCFS_Counts_Summary_fig.png")
# plt.setp(label, fontsize=13)
plt.show()
```
